```python
import os
import shutil
import random

# Definir las rutas
base_dir = '/content/dataset/Juegos completos imagenes y etiquetas/Juegos completos imagenes y etiquetas'
train_images_dir = '/content/dataset/images/train'
val_images_dir = '/content/dataset/images/val'
train_labels_dir = '/content/dataset/labels/train'
val_labels_dir = '/content/dataset/labels/val'

# Crear las carpetas necesarias
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# Listar todos los archivos
all_files = os.listdir(base_dir)

# Filtrar imágenes y etiquetas
images = [f for f in all_files if f.endswith('.jpg') or f.endswith('.png')]
labels = [f for f in all_files if f.endswith('.txt')]

# Crear conjuntos de nombres de archivos para validación
image_labels_map = {img.split('.')[0]: img for img in images}
print("Mapeo de imágenes a etiquetas:", image_labels_map)

label_names = set(label.split('.')[0] for label in labels)
print("Nombres de etiquetas disponibles:", label_names)

# Filtrar las imágenes y etiquetas para mantener la correspondencia
valid_images = [img for img in images if img.split('.')[0] in label_names]
valid_labels = [lbl for lbl in labels if lbl.split('.')[0] in image_labels_map]
print("Imágenes válidas:", valid_images)
print("Etiquetas válidas:", valid_labels)

# Mezclar aleatoriamente las imágenes y etiquetas
random.shuffle(valid_images)
random.shuffle(valid_labels)

# Dividir en entrenamiento y validación (80% entrenamiento, 20% validación)
split_ratio = 0.8
num_train_images = int(len(valid_images) * split_ratio)
num_train_labels = int(len(valid_labels) * split_ratio)

train_images = valid_images[:num_train_images]
val_images = valid_images[num_train_images:]
train_labels = valid_labels[:num_train_labels]
val_labels = valid_labels[num_train_labels:]

# Mover imágenes y etiquetas a las carpetas correspondientes
for image in train_images:
    label = image.split('.')[0] + '.txt'
    if os.path.exists(os.path.join(base_dir, label)):
        print(f"Moviendo etiqueta a train: {label}")
        shutil.move(os.path.join(base_dir, label), os.path.join(train_labels_dir, label))
    else:
        print(f"Etiqueta no encontrada para {image}: {label}")
    shutil.move(os.path.join(base_dir, image), os.path.join(train_images_dir, image))

for image in val_images:
    label = image.split('.')[0] + '.txt'
    if os.path.exists(os.path.join(base_dir, label)):
        print(f"Moviendo etiqueta a val: {label}")
        shutil.move(os.path.join(base_dir, label), os.path.join(val_labels_dir, label))
    else:
        print(f"Etiqueta no encontrada para {image}: {label}")
    shutil.move(os.path.join(base_dir, image), os.path.join(val_images_dir, image))

# Verificar el contenido después de mover
print("Archivos en train/images:", os.listdir(train_images_dir))
print("Archivos en val/images:", os.listdir(val_images_dir))
print("Archivos en train/labels:", os.listdir(train_labels_dir))
print("Archivos en val/labels:", os.listdir(val_labels_dir))

```

```python

from ultralytics import YOLO
from google.colab import files
import os
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

# Subir el archivo del modelo
print("Sube tu archivo de modelo (.pt):")
uploaded_model = files.upload()

# Obtener el nombre del archivo del modelo
model_path = list(uploaded_model.keys())[0]
print(f"Modelo cargado: {model_path}")

# Cargar el modelo
model = YOLO(model_path)

# Subir la imagen
print("Sube la imagen que deseas analizar:")
uploaded_image = files.upload()

# Obtener el nombre del archivo de la imagen
image_path = list(uploaded_image.keys())[0]
print(f"Imagen cargada: {image_path}")

# Realizar la predicción
results = model(image_path)

# Iterar sobre los resultados y analizar las detecciones
for result in results:
    # Obtener las detecciones
    boxes = result.boxes  # Objeto con las cajas detectadas

    # Si quieres obtener los datos en formato NumPy:
    detections = boxes.data.cpu().numpy()



    # Analizar cada detección
    for detection in detections:
        x1, y1, x2, y2, confidence, class_id = detection
        class_name = model.names[int(class_id)]
        print(f"Detección: {class_name}, Confianza: {confidence:.2f}")

    # Determinar si se detectó contenido violento
    if any(int(detection[5]) == 1 for detection in detections):  # Ajusta el ID según corresponda
        print("¡Alerta! Juego violento detectado.")
    else:
        print("Juego apropiado.")

    # Mostrar y guardar la imagen con las detecciones
    annotated_image = result.plot()
    output_image_path = 'resultado.jpg'
    cv2.imwrite(output_image_path, annotated_image)
    print(f"Imagen con detecciones guardada en {output_image_path}")

    # Mostrar la imagen dentro de Colab
    img = mpimg.imread(output_image_path)
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
