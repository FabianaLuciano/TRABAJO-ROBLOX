```python
import pandas as pd

# Suponiendo que el archivo subido se llama 'simonazosimon - Roblox.csv'
csv_filename = 'simonazosimon - Roblox.csv'

# Cargar el archivo CSV con el historial de juegos
df = pd.read_csv(csv_filename)

# Mostrar las primeras filas del DataFrame para verificar la carga correcta
print(df.head())
```

```python
# Preprocesamiento del texto (ajustar seg√∫n sea necesario)
def preprocess_text(text):
    # Implementa el preprocesamiento necesario para tu modelo
    return text

# Agregar una columna para los resultados
df['es_violento'] = ''

# Procesar cada juego
for index, row in df.iterrows():
    game_name = row['T√≠tulo']  # Nombre del juego
    game_text = row['Texto']  # Descripci√≥n o texto adicional del juego

    # Combinar el t√≠tulo y el texto en un solo campo para la predicci√≥n
    full_text = preprocess_text(game_name + " " + game_text)  # Ajustar el preprocesamiento seg√∫n sea necesario

    # Realizar la predicci√≥n
    try:
        prediction = model.predict([full_text])[0]
    except Exception as e:
        print(f"Error al predecir el juego '{game_name}': {e}")
        prediction = None  # O alg√∫n valor por defecto

    # Decidir el mensaje basado en la predicci√≥n
    if prediction == 1:
        result = "Este juego es violento"
    elif prediction == 0:
        result = "Este juego parece apropiado"
    else:
        result = "Predicci√≥n no v√°lida"

    # Guardar el resultado en el DataFrame
    df.at[index, 'Clase'] = result

# Guardar los resultados en un nuevo archivo CSV en el directorio actual
output_filename = '/content/historial_clasificado.csv'
df.to_csv(output_filename, index=False)

print(f"Archivo guardado en: {output_filename}")
```
