```python
from google.colab import files
from PIL import Image
import os

# Paso 1: Subir archivos .webp
uploaded = files.upload()

# Crear un directorio para guardar los archivos .jpg
os.makedirs('converted_images', exist_ok=True)

# Paso 2: Convertir los archivos .webp a .jpg
for filename in uploaded.keys():
    if filename.lower().endswith('.webp'):
        # Cargar la imagen .webp
        image = Image.open(filename)

        # Convertir a .jpg y guardar
        jpg_filename = f"converted_images/{os.path.splitext(filename)[0]}.jpg"
        image.convert('RGB').save(jpg_filename, 'JPEG')
        print(f"Convertido: {filename} a {jpg_filename}")

# Paso 3: Descargar los archivos convertidos
import shutil
shutil.make_archive('converted_images', 'zip', 'converted_images')
files.download('converted_images.zip')
```
