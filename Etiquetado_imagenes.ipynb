```python
from PIL import Image
import matplotlib.pyplot as plt

# Mostrar algunas im√°genes
def show_images(image_folder, num_images=5):
    image_files = os.listdir(image_folder)
    for i, img_file in enumerate(image_files[:num_images]):
        img_path = os.path.join(image_folder, img_file)
        img = Image.open(img_path)
        plt.subplot(1, num_images, i+1)
        plt.imshow(img)
        plt.title(img_file)
        plt.axis('off')
    plt.show()

show_images(image_folder, num_images=5)
```

```python
import os

# Carpeta con im√°genes
image_folder = '/content/imagenes_y_etiquetas_apropiadas/Imagenes apropiadas en jpg'
# Carpeta para guardar etiquetas
label_folder = '/content/imagenes/labels'
os.makedirs(label_folder, exist_ok=True)

# ID de la clase "apropiado" en YOLO
class_id = 0

# Crear archivos de etiquetas para cada imagen
for img_file in os.listdir(image_folder):
    if img_file.endswith('.jpg'):  # Aseg√∫rate de ajustar la extensi√≥n si es necesario
        # Crear el archivo de etiqueta correspondiente
        label_file = os.path.splitext(img_file)[0] + '.txt'
        label_path = os.path.join(label_folder, label_file)

        # Escribir la etiqueta en el archivo
        with open(label_path, 'w') as f:
            # Suponiendo que cada imagen tiene una sola etiqueta con coordenadas dummy
            # En realidad, deber√≠as ajustar esto con las coordenadas reales del objeto
            f.write(f"{class_id} 0.5 0.5 1.0 1.0\n")  # Coordenadas dummy: (x_center, y_center, width, height)

print("Archivos de etiqueta creados en:", label_folder)
```

```python
import zipfile
import os

# Directorio que quieres comprimir
folders_to_zip = ['/content/imagenes', '/content/imagenes/labels']

# Nombre del archivo ZIP
zip_filename = '/content/imagenes_y_labels.apropiadas_comprimidas.zip'

# Crear un archivo ZIP con las carpetas especificadas
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for folder in folders_to_zip:
        for root, dirs, files in os.walk(folder):
            for file in files:
                # Crear la ruta del archivo en el archivo ZIP
                file_path = os.path.join(root, file)
                # Ruta relativa del archivo dentro del archivo ZIP
                relative_path = os.path.relpath(file_path, os.path.dirname(folder))
                zipf.write(file_path, relative_path)

print(f"Archivo ZIP creado: {zip_filename}")
```

```python
import zipfile
import os

# Directorios de im√°genes y etiquetas
image_folder = '/content/imagenes_y_etiquetas_apropiadas/Imagenes apropiadas en jpg'
label_folder = '/content/imagenes/labels'

# Ruta del archivo ZIP a crear
zip_filename = '/content/imagenes_y_etiquetas_apropiadas_comprimidas.zip'

# Crear un archivo ZIP que incluya tanto las im√°genes como las etiquetas
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    # A√±adir im√°genes al ZIP
    for root, dirs, files in os.walk(image_folder):
        for file in files:
            if file.endswith('.jpg'):
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, image_folder))

    # A√±adir etiquetas al ZIP
    for root, dirs, files in os.walk(label_folder):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, label_folder))

print(f"Archivo ZIP creado: {zip_filename}")

```
